```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import os
from os.path import join
```

```python
def str_to_path(path_str):
    '''Converts windows path string to os.path'''
    path_list = path_str.split('\\')
    path = os.path.join(path_list[0], os.sep, *path_list[1:])
    return path
```

```python
path = str_to_path(r"N:\Projects\Groundwater\projects\Global_GDE_Map\data")
plots_in = os.path.join(path, "sPlotOpen", "sPlotOpen_header(2).txt")
veg_in = os.path.join(path, "sPlotOpen", "sPlotOpen_DT(1).txt")
```

```python
df1 = pd.read_csv(plots_in, sep='\t')
df1
```

```python
df2 = pd.read_csv(veg_in, sep='\t')
df2
```

```python
df3 = pd.merge(df2, df1, on='PlotObservationID', how='left')
df4 = (df3.groupby(['Species', 'Continent'], dropna=False)
           .agg({'PlotObservationID':'count'})
           .reset_index()
           .rename(columns={'PlotObservationID':'Count_plots'})
          )
df4
```

```python
out_csv = os.path.join(path, "sPlotOpen_species_continent_table.csv")
df4.to_csv(out_csv, encoding='utf-8')
```

```python
df5 = df4.pivot_table(index=['Species'], columns='Continent', values='Count_plots').reset_index()
df5
```

```python
# Export genus level distribution tables
genus = 'Salix'
out_csv = os.path.join(path, "sPlotOpen_genus_continent_table_for_{}.csv".format(genus))
df6 = df5[df5['Species'].str.startswith(genus)]
df6.to_csv(out_csv, encoding='utf-8')
df6
```

```python
# Export country level species lists
country = 'Mongolia'
out_csv = os.path.join(path, "sPlotOpen_country_species_table_for_{}.csv".format(country))
df7 = df3.loc[df3['Country'] == country]
df8 = (df7.groupby(['Species', 'Country'], dropna=False)
           .agg({'PlotObservationID':'count'})
           .reset_index()
           .rename(columns={'PlotObservationID':'Count_plots'})
          )
df8.to_csv(out_csv, encoding='utf-8')
df8
```
