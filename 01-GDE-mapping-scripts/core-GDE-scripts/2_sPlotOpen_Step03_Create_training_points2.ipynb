```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import os
from os.path import join
```

```python
def str_to_path(path_str):
    '''Converts windows path string to os.path'''
    path_list = path_str.split('\\')
    path = os.path.join(path_list[0], os.sep, *path_list[1:])
    return path
```

```python
path = str_to_path(r"N:\WATER_Program\Groundwater\Global_GDE_Map\data")
plots_in = os.path.join(path, "sPlotOpen", "sPlotOpen_header(2).txt")
veg_in = os.path.join(path, "sPlotOpen", "sPlotOpen_DT(1).txt")
tbl_in = str_to_path(r"N:\WATER_Program\Groundwater\Global_GDE_Map\tables\sPlotOpen_GDE_review_20230407.csv")
tbl_in2 = str_to_path(r"N:\WATER_Program\Groundwater\Global_GDE_Map\tables\sPlotOpen_GDE_review_20230407_forToderichReview_Kristina Last_KK.xlsx")
tbl_out = str_to_path(r"N:\WATER_Program\Groundwater\Global_GDE_Map\tables\sPlotOpen_GDE_points_20230418.csv")
tbl_out2 = str_to_path(r"N:\WATER_Program\Groundwater\Global_GDE_Map\tables\sPlotOpen_GDE_review_20230418.csv")
```
